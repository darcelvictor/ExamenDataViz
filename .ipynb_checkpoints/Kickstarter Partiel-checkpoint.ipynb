```python
df['state'].value_counts(sort="F").plot.pie(figsize=(20, 20))
```

```python
df1=pd.value_counts(df["country"])
df1.plot.bar(rot=0,figsize=(20, 10),title='Nombre de projets pas pays',fontsize=20)
```

```python
df1.plot.bar(rot=0,figsize=(20, 10),title='Nombre de projets pas pays echelle logarithmique',fontsize=20,logy=True)
```

```python
df2=df[df['state']=='successful'].groupby("country").usd_goal_real.mean()
df2.plot.bar(rot=0,figsize=(20, 10),title='Objectif moyen des projets réussis par pays',fontsize=20)
```

```python
df3=df[df['state']=='successful'].groupby("country").usd_pledged_real.mean()
df3.plot.bar(rot=0,figsize=(20, 10),title='Montant moyen obtenue des projets réussis par pays',fontsize=20)
```

```python
df4= pd.DataFrame({'usd_goal_mean': df2,'usd_pledged_real_mean': df3}, index=df3.index)
df4.plot.bar(rot=0,figsize=(20, 10),title="Tableau comparant les moyennes des objectifs et des sommes obtenues par pays pour les projets réussis")
```

```python
df6=df[df['state']=='successful'].groupby("category").usd_goal_real.mean()
df6.sort_values(ascending=True).plot.barh(rot=0,figsize=(20, 100),title='Objectif moyen en dollars des projets réussis par categotie',fontsize=20)
```

```python
df['deadline'] = pd.to_datetime(df.deadline, format='%Y-%m-%d')
df['launched'] = pd.to_datetime(df.launched, format='%Y-%m-%d %H:%M:%S')
df['diff_days'] = df['deadline'] - df['launched']
df['diff_days']=df['diff_days']/np.timedelta64(1,'D')

df
```
