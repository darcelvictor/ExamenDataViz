# Etude du temps par projet

Cr√©ation d'un nouvelle colonne

```python
df['deadline'] = pd.to_datetime(df.deadline, format='%Y-%m-%d')
df['launched'] = pd.to_datetime(df.launched, format='%Y-%m-%d %H:%M:%S')
df['diff_days'] = df['deadline'] - df['launched']
df['diff_days']=df['diff_days']/np.timedelta64(1,'D')

df.head()
```

```python
print("Delais moyen pour les projets qui ont echou√©s",df[df["state"]=="failed"].diff_days.mean())
print("Delais moyen pour les projets qui ont r√©ussis",df[df["state"]=="successful"].diff_days.mean())
```

```python
df3=df[df["state"]=="failed"].groupby("category").diff_days.mean()-df[df["state"]=="successful"].groupby("category").diff_days.mean()
```

# Don Moyen par projet

```python
df['arevage_donation'] = df['usd_pledged_real'] /df['backers']
df['arevage_donation'].fillna(0, inplace=True)
df
```

## Donation median par pays

```python
df2=df.groupby("country").arevage_donation.median()
df2.plot.bar(figsize=(20, 10),title='Donation median par pays',fontsize=20)
```

## Donation median par cat√©gories

```python
df3=df.groupby("category").arevage_donation.median()
df3.plot.barh(figsize=(20, 100),title='Donation median par category',fontsize=20)
```

## Donation median par mois 

```python
df3=df.groupby(df.launched.dt.month).arevage_donation.median()
df3.plot.bar(figsize=(20, 20),title='Donation median en fonction des mois',fontsize=20)
```

on se rend compte que les gens d√©pensent moins l'√©t√© et fin d'ann√©e

```python
df3=df.groupby(df.launched.dt.year).arevage_donation.median()
df3.plot.bar(figsize=(20, 20),title='Donation median en fonction des mois',fontsize=20)
```
